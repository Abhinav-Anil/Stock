```
#using statmodel package
import statsmodels.api as sm

import pandas.core.tools.datetimes as datetools
```

```
#inaliizing short and long windows
long_window = 100
short_window = 40
```

```
#Intialize the signal with the signal column 
signals = pd.DataFrame(index = spx.index)
signals['signal'] = 0.0
```

```
#create short simple moving average over the short window
signals['short_mavg'] = spx['Close'].rolling(window = short_window, min_periods =1 , center=False).mean()
```

```
#create long simple moving average over the long window
signals['long_mavg'] = spx['Close'].rolling(window = long_window, min_periods =1 , center = False).mean()
```

```
# Create signals
signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] > signals['long_mavg'][short_window:], 1.0, 0.0)
```

```
#Generate trading  orders
signals['positions'] = signals['signal'].diff()
```

```
#signal output
print(signals[20:])
```

```
# Initialize the plot figure
fig = plt.figure(figsize=(18,8))

# Add a subplot and label for y-axis
ax1 = fig.add_subplot(111,  ylabel='Price in $')

# Plot the closing price
spx['Close'].plot(ax=ax1, color='r', lw=2.)

# Plot the short and long moving averages
signals[['short_mavg', 'long_mavg']].plot(ax=ax1, lw=2.)

# Plot the buy signals
ax1.plot(signals.loc[signals.positions == 1.0].index, 
         signals.short_mavg[signals.positions == 1.0],
         '^', markersize=10, color='green')
         
# Plot the sell signals
ax1.plot(signals.loc[signals.positions == -1.0].index, 
         signals.short_mavg[signals.positions == -1.0],
         'v', markersize=10, color='blue')
         
# Show the plot
plt.show()
```

```
#back testing
# Set the initial capital
initial_capital= float(100000.0)

# Create a DataFrame `positions`
positions = pd.DataFrame(index=signals.index).fillna(0.0)

# Buy a 100 shares
positions['SPX'] = 100*signals['signal']   
  
# Initialize the portfolio with value owned   
portfolio = positions.multiply(spx['Close'], axis=0)

# Store the difference in shares owned 
pos_diff = positions.diff()

# Add `holdings` to portfolio
portfolio['holdings'] = (positions.multiply(spx['Close'], axis=0)).sum(axis=1)

# Add `cash` to portfolio
portfolio['cash'] = initial_capital - (pos_diff.multiply(spx['Close'], axis=0)).sum(axis=1).cumsum()   

# Add `total` to portfolio
portfolio['total'] = portfolio['cash'] + portfolio['holdings']

# Add `returns` to portfolio
portfolio['returns'] = portfolio['total'].pct_change()

# Print the first lines of `portfolio`
print(portfolio.head())
```

```
# Create a figure
fig = plt.figure(figsize=(20,10))

ax1 = fig.add_subplot(111, ylabel='Portfolio value in $')

# Plot the equity curve in dollars
portfolio['total'].plot(ax=ax1, lw=2.)

ax1.plot(portfolio.loc[signals.positions == 1.0].index, 
         portfolio.total[signals.positions == 1.0],
         '^', markersize=10, color='m')
ax1.plot(portfolio.loc[signals.positions == -1.0].index, 
         portfolio.total[signals.positions == -1.0],
         'v', markersize=10, color='k')

# Show the plot
plt.show()
```

```
# Define a trailing 60 trading day window
window = 60

# Calculate the max drawdown in the past window days for each day 
rolling_max = spx['Close'].rolling(window, min_periods=1).max()
daily_drawdown = spx['Close']/rolling_max - 1.0

# Calculate the minimum (negative) daily drawdown
max_daily_drawdown = daily_drawdown.rolling(window, min_periods=1).min()

# Plot the results
daily_drawdown.plot()
max_daily_drawdown.plot()

# Show the plot
plt.show()
```
